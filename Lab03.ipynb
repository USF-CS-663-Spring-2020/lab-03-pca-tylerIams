# **Part 1** - Breast Cancer Data

```python
import pandas as pd
```

```python
df = pd.read_csv('wdbc.data.csv', header=None)
```

```python
df
```

Let's apply labels to the columns

```python
column_labels = ['patient_id', 'diagnosis', 'mean radius', 'mean texture', 'mean perimeter', 'mean area',
'mean smoothness', 'mean compactness', 'mean concavity',
'mean concave points', 'mean symmetry', 'mean fractal dimension',
'radius error', 'texture error', 'perimeter error', 'area error',
'smoothness error', 'compactness error', 'concavity error',
'concave points error', 'symmetry error',
'fractal dimension error', 'worst radius', 'worst texture',
'worst perimeter', 'worst area', 'worst smoothness',
'worst compactness', 'worst concavity', 'worst concave points',
'worst symmetry', 'worst fractal dimension']
```

```python
df.columns = column_labels
df
```

```python
df['diagnosis'].replace('B', 'Benign', inplace=True)
df['diagnosis'].replace('M', 'Malignant', inplace=True)
```

```python
df
```

See how many unique and null values we have:

```python
df.diagnosis.unique()
```

```python
df.diagnosis.isnull().sum()
```

```python
from sklearn.preprocessing import *
```

Separate the features from the labels into two dataframes

```python
feature_df = df.iloc[:,2:]
```

```python
label_df = df.iloc[:,1]
```

```python
feature_df
```

```python
label_df
```

Normalize the data by taking each value's z-score

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(feature_df)
```

```python
x
```

Check out the mean and standard deviation of the normalized data -- we can see that the mean is (basically) zero and the standard deviation is 1.0, which indicates a normal distribtion.

```python
import numpy as np
np.mean(x), np.std(x)
```

Apply a PCA

```python
from sklearn.decomposition import PCA

pca_breast = PCA(n_components=4)
```

```python
principalComponents_breast = pca_breast.fit_transform(x)
```

```python
pca_df = pd.DataFrame(principalComponents_breast, columns = ['PC1', 'PC2','PC3','PC4'])
```

We can see (since we chose to use 4 principal components above) we have transformed the data into an nx4 matrix using our 4 principal components

```python
pca_df
```

We can see that our first principal component (PC1) explains ~44% of the data, PC2 explains ~19%, PC3 ~9% and PC4 ~7%

```python
pca_breast.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Now let's plot the first two principal components and color by their class (Benign v Malignant) to see if we can spot a relationship

```python
plt.figure()
plt.xlabel('PC1')
plt.ylabel('PC2')
targets = ['Benign', 'Malignant']
colors = ['b', 'r']
for target, color in zip(targets, colors):
    indicesToKeep = label_df == target
    plt.scatter(pca_df.loc[indicesToKeep, 'PC1'],
                pca_df.loc[indicesToKeep, 'PC2'],
                c = color)
```

It's pretty clear that there is a relationship between instances and classes that can be easily visualized using only the first two principal components!  These two together explain about 63% of the variance in the data, but are still a good representation of the classification problem we might like to solve later.

```python
pca = PCA().fit(x)
evr = pca.explained_variance_ratio_
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(evr)))
```

Let's plot the cumulative variance explained by adding more components.

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

We can see that if we used a little more than 5 PC's we could explain about 90% of the variance.  However, we saw that we really only needed 2 principal components to (seemingly) sufficiently classify into Malignant and Benign in our potential classification problem.

Let's now normalize the data by taking each value's z-score

```python
gi_norm = StandardScaler().fit_transform(gi_feat)
np.mean(gi_norm), np.std(gi_norm)
```

We can again see that our mean is basically zero and our standard deviation is about 1, which indicate a normal distribution

This data has way more features than our other dataset (5000 vs 32) so we will take more principal comonents due to the higher dimensionailty (which will increase the overall distance between our data in multi-dimensional space) and will thus have higher variance from any single PC.  

```python
pca_gi = PCA(n_components=10)
```

```python
pca_gisette = pca_gi.fit_transform(gi_norm)
```

Let's put our PCA into a dataframe

```python
df_pca_gisette = pd.DataFrame(pca_gisette, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 'PC10'])
```

We can visualize our PCA dataframe below (although the numbers aren't meaningful to a human eye, they represent the data as projected from it's full dimensional space into our 10 dimensional space by our PCA)

```python
df_pca_gisette
```

Let's look at how much variance each principal component explains

```python
pca_gi.explained_variance_ratio_
```

We can see that principal components 1 and 2 explain about 3% and 2% of the variance, respectively. Components 3 and 4 each explain about 1.6% of the variance, and components 5-8 each explain around 1% of the variance.  None of them capture nearly as much variance as we saw with our previous PCA's components.  This makes sense though given how many more dimensions this data exists in and how much multidimensional space is thus between each individual datapoint whithin that larger space.

Now let's look at the cumulative PCA component graph to see if we can find out how many components might begin to explain a significant amount of our data

```python
gipca = PCA().fit(gi_norm)
```

```python
gi_evr = gipca.explained_variance_ratio_
#Let's only see the first 50 and multiply the numbers by 100 to more easily conceptualize what they're doing
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(gi_evr[:50] * 100))) 
```

It looks like they're explaining less than a half a percent each after the first 12 or so.  Let's plot this to try to make sense of it

```python
plt.plot(np.cumsum(gipca.explained_variance_ratio_))
```

It looks like we'd need around 1200 principal components before they capture about 75% of the variance.  Which would significantly reduce our dimensionality in practice!
